QUESTION 3:Write a simple calculator program using match case (addition, subtraction, multiplication, division).

```python
# Simple Calculator using match-case

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Choose operation:")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")

choice = input("Enter operation (+, -, *, /): ")

match choice:
    case "+":
        print(f"Result: {num1} + {num2} = {num1 + num2}")
    case "-":
        print(f"Result: {num1} - {num2} = {num1 - num2}")
    case "*":
        print(f"Result: {num1} * {num2} = {num1 * num2}")
    case "/":
        if num2 != 0:
            print(f"Result: {num1} / {num2} = {num1 / num2}")
        else:
            print("Error! Division by zero ")
    case _:
        print("Invalid operation ")
```

Task 4: While Loop

QUESTION 1:Write a program to print numbers from 1 to 10 using while.

```python
i = 1  # starting number
while i <= 10:
    print(i)
    i += 1  # increase i by 1
```

QUESTION 2:Write a program to compute the sum of first N natural numbers using while.

```python
N = int(input("Enter a number (N): "))

i = 1
total = 0

while i <= N:
    total += i   # add current number
    i += 1       # move to next number

print(f"Sum of first {N} natural numbers is: {total}")
```

QUESTION 3:Write a program to reverse the digits of a given number (e.g., 123 → 321).



```python
num = int(input("Enter a number: "))

reverse = 0
while num > 0:
    digit = num % 10        # get last digit
    reverse = reverse * 10 + digit  # add digit to reverse
    num = num // 10         # remove last digit

print("Reversed number:", reverse)
```

QUESTION 1:Write a program to print the multiplication table of a number.

```python
num = int(input("Enter a number: "))

print(f"\nMultiplication Table of {num}:")
i = 1
while i <= 10:
    print(f"{num} x {i} = {num * i}")
    i += 1
```

QUESTION 2:Write a program to find the factorial of a given number.

```python
num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print(f"Factorial of {num} is: {factorial}")
```

QUESTION 3:Write a program to count the number of vowels in a string.

```python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"
count = 0

for char in text:
    if char in vowels:
        count += 1

print(f"Number of vowels in the string: {count}")
```
